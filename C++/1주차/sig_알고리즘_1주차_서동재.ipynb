```python
# 10828
import sys
input = sys.stdin.readline

N = int(input())
num_list = []
for i in range(N):
  new_list = list(input().split())
  if new_list[0] == 'push':
    if len(num_list) >= 1:
      temp_list = []
      temp_list = [int(new_list[1])] + num_list
      num_list = temp_list
    else:
      num_list.append(int(new_list[1]))
  elif new_list[0] == 'pop':
    if len(num_list) >= 1:
      print(num_list[0])
      temp_list = num_list[1:]
      num_list = temp_list
    else :
      print(-1)
  elif new_list[0] == 'size':
    print(len(num_list))
  elif new_list[0] == 'empty':
    if len(num_list) >= 1:
      print(0)
    else :
      print(1)
  elif new_list[0] == 'top':
    if len(num_list) >= 1:
      print(num_list[0])
    else:
      print(-1)
```

```python
#1874
import sys
input = sys.stdin.readline

n = int(input())
target = []
for i in range(n):
  target.append(int(input()))

stack = []
num = 1
check_list = []
try:
  for i in target:
    while num < i :
      stack.append(num)
      num = num + 1
      check_list.append("+")

    if num == i :
      stack.append(num)
      check_list.append("+")
      del stack[-1]
      check_list.append("-")
      num = num + 1
      continue

    while num > i:
      if stack[-1] == i:
        del stack[-1]
        check_list.append("-")
        break
      else:
        del stack[-1]
        check_list.append("-")
except:
  check_list.append("NO")

if check_list[-1] == "NO":
  print("NO")
else:
  for i in check_list:
    print(i)
```

```python
#1406
import sys
input = sys.stdin.readline

S = list(input().strip())
N = len(S)
M = int(input())
stack = []
for i in range(M):
  new_list = list(input().split())
  if new_list[0] == 'L' and S != []:
    stack.append(S.pop())
  elif new_list[0] == 'D' and stack != []:
    S.append(stack.pop())
  elif new_list[0] == 'B' and S != []:  
    S.pop()
  elif new_list[0] == 'P':
    S.append(new_list[1])

print("".join(S + list(reversed(stack))))
```

```python
#10866
import sys
input = sys.stdin.readline

N = int(input())
num_list = []
for i in range(N):
  new_list = list(input().split())
  if new_list[0] == 'push_back':
    num_list.append(int(new_list[1]))
  elif new_list[0] == 'push_front':
    num_list.insert(0, int(new_list[1]))
  elif new_list[0] == 'pop_front':
    if len(num_list) >= 1:
      print(num_list[0])
      del num_list[0]
    else :
      print(-1)
  elif new_list[0] == 'pop_back':
    if len(num_list) >= 1:
      print(num_list[-1])
      del num_list[-1]
    else :
      print(-1)
  elif new_list[0] == 'size':
    print(len(num_list))
  elif new_list[0] == 'empty':
    if len(num_list) >= 1:
      print(0)
    else :
      print(1)
  elif new_list[0] == 'front':
    if len(num_list) >= 1:
      print(num_list[0])
    else :
      print(-1)
  elif new_list[0] == 'back':
    if len(num_list) >= 1:
      print(num_list[-1])
    else :
      print(-1)
```

```python
#17298
n = int(input())
A = list(map(int, input().split()))
answer = [-1] * n
stack = []

stack.append(0)
for i in range(1, n):
    while stack and A[stack[-1]] < A[i]:
        answer[stack.pop()] = A[i]
    stack.append(i)

print(*answer)
```

```python
#17299
import sys
input = sys.stdin.readline

n = int(input())
A = list(map(int, input().split()))
answer = [-1] * n
stack = []
check_list = [0] * 1000001
checking = [0] * n
for i in A:
  check_list[i] = check_list[i] + 1

for i in range(len(A)):
  checking[i] = check_list[A[i]]

stack.append(0)
for i in range(1, n):
    while stack and checking[stack[-1]] < checking[i]:
        answer[stack.pop()] = A[i]
    stack.append(i)

print(*answer)
```
